## 1. Preprocessing Steps

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# è‹¥å·²ä¸‹è¼‰éå¯è¨»è§£æ‰
nltk.download('stopwords')
nltk.download('punkt')

stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text) # å»ç¶²å€
    text = re.sub(r'\@\w+|\#', '', text)                 # å» @user, #hashtag
    text = text.translate(str.maketrans('', '', string.punctuation)) # å»æ¨™é»
    text = re.sub(r'\d+', '', text)                      # å»æ•¸å­—
    
    tokens = text.split()
    tokens = [stemmer.stem(w) for w in tokens if w not in stop_words]
    return " ".join(tokens)

print("Start preprocessing...")
# ä½ çš„è³‡æ–™æ¬„ä½å« 'content'ï¼Œé€™è£¡å°æ‡‰ä¿®æ­£
df_train['clean_content'] = df_train['content'].apply(preprocess_text)
df_test['clean_content'] = df_test['content'].apply(preprocess_text)
print("Preprocessing done! Cleaned content created.")
```

```python
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, classification_report

print("å•Ÿå‹•é«˜éšæ¨¡å¼ï¼šWord + Char ç‰¹å¾µèåˆæ¨¡å‹...")

# 1. å®šç¾©ç‰¹å¾µæå–å™¨ (Feature Union)
# çµåˆã€Œå–®å­—ã€å’Œã€Œå­—å…ƒã€å…©ç¨®è§€é»
combined_features = FeatureUnion([
    # è§€é» A: çœ‹å–®å­— (Word) - æŠ“é—œéµæ„æ¶µ
    ('word_tfidf', TfidfVectorizer(
        ngram_range=(1, 2),      # çœ‹å–®å­—å’Œé›™å­—è©
        max_features=20000,      # ä¿ç•™æ›´å¤šç‰¹å¾µ
        min_df=2,
        analyzer='word',
        stop_words=None          # å‰é¢æˆ‘å€‘å·²ç¶“è™•ç†éåœç”¨è©ï¼Œé€™è£¡è¨­ None é¿å…é‡è¤‡
    )),
    
    # è§€é» B: çœ‹å­—å…ƒ (Char) - æŠ“æ‹¼å¯«ç¿’æ…£èˆ‡èªæ°£ (å¦‚: "sooooo", "!!!")
    ('char_tfidf', TfidfVectorizer(
        ngram_range=(3, 5),      # æŠ“ 3 åˆ° 5 å€‹å­—æ¯çš„ç‰‡æ®µ
        max_features=10000,
        analyzer='char_wb',      # char_wb æœƒè€ƒæ…®å–®å­—é‚Šç•Œï¼Œæ•ˆæœé€šå¸¸æ›´å¥½
        min_df=2
    )),
])

# 2. å®šç¾©æ¨¡å‹çµ„åˆ (Ensemble)
# çµåˆ Logistic Regression å’Œ MultinomialNB
# (LinearSVC ä¸æ”¯æ´ soft votingï¼Œé€™è£¡å…ˆç”¨é€™å…©å€‹æ©Ÿç‡æ¨¡å‹)
clf1 = LogisticRegression(max_iter=2000, class_weight='balanced', solver='lbfgs', C=1.0)
clf2 = MultinomialNB(alpha=0.1) # Naive Bayes å°æ–‡å­—åˆ†é¡å¸¸æœ‰å¥‡æ•ˆ

ensemble_clf = VotingClassifier(
    estimators=[('lr', clf1), ('nb', clf2)],
    voting='soft', # ä½¿ç”¨æ©Ÿç‡åŠ æ¬ŠæŠ•ç¥¨
    weights=[1.5, 1] # ç¨å¾®åŠ é‡ LR çš„æ¬Šé‡ (ç¶“é©—æ³•å‰‡)
)

# 3. å»ºç«‹ç®¡ç·š (Pipeline)
pipeline = Pipeline([
    ('features', combined_features),
    ('classifier', ensemble_clf)
])

# 4. æº–å‚™è³‡æ–™
# æ³¨æ„ï¼šé€™è£¡ç›´æ¥ç”¨ df_train['clean_content'] (é€™æ˜¯æ–‡å­—)ï¼ŒPipeline æœƒè‡ªå·±åš TF-IDF
X_input = df_train['clean_content'].astype(str)
y_input = df_train['sentiment']

X_tr, X_val, y_tr, y_val = train_test_split(X_input, y_input, test_size=0.1, random_state=42)

# 5. è¨“ç·´èˆ‡é©—è­‰
print("æ­£åœ¨è¨“ç·´ Feature Union æ¨¡å‹ (æœƒæ¯”è¼ƒä¹…ï¼Œè«‹ç¨ç­‰)...")
pipeline.fit(X_tr, y_tr)

val_preds = pipeline.predict(X_val)
val_f1 = f1_score(y_val, val_preds, average='macro')
print(f"Validation Macro F1 Score: {val_f1:.4f}")
print(classification_report(y_val, val_preds))

# --- å…¨è³‡æ–™é‡æ–°è¨“ç·´èˆ‡æäº¤ ---
print("æ­£åœ¨ä½¿ç”¨å…¨è³‡æ–™é‡æ–°è¨“ç·´...")
pipeline.fit(X_input, y_input)
final_preds = pipeline.predict(df_test['clean_content'].astype(str))

submission = pd.DataFrame({
    'id': df_test['id'],
    'emotion': final_preds
})
submission.to_csv('submission_ensemble_char.csv', index=False)
print("submission_ensemble_char.csv å·²ç”Ÿæˆï¼é€™æ‹›é€šå¸¸èƒ½çªç ´ 0.6ï¼")
```

```python
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, classification_report

print("Feature Union + ç´” SVM")

# 1. ç‰¹å¾µæå– (ç¨å¾®ç¸®å°ç¯„åœï¼Œæ¸›å°‘é›œè¨Š)
combined_features = FeatureUnion([
    # Word: åªçœ‹ 1-2 gram (ç‰‡èª)ï¼Œä¸è¦ 3 gram (å¤ªé•·å®¹æ˜“éæ“¬åˆ)
    ('word_tfidf', TfidfVectorizer(
        ngram_range=(1, 2),
        max_features=20000,
        min_df=2,
        analyzer='word',
        sublinear_tf=True,  # å¹³æ»‘åŒ–ä¿ç•™
        stop_words=None     # ä½¿ç”¨æˆ‘å€‘è‡ªå®šç¾©çš„ä¿ç•™å¦å®šè©ç‰ˆæœ¬
    )),
    
    # Char: æ”¹å› 2-4 gram (å°ˆæ³¨æ–¼å­—æ ¹å’Œè¡¨æƒ…ç¬¦è™Ÿç‰¹å¾µ)
    ('char_tfidf', TfidfVectorizer(
        ngram_range=(2, 4),
        max_features=15000,
        analyzer='char_wb',
        sublinear_tf=True,
        min_df=2
    )),
])

# 2. æ¨¡å‹ï¼šåªç”¨ LinearSVC
# C=0.8: ç¨å¾®é™ä½æ­£è¦åŒ–å¼·åº¦ (é è¨­ 1.0)ï¼Œè®“æ¨¡å‹å®¹éŒ¯ç‡é«˜ä¸€é»é»
model = LinearSVC(class_weight='balanced', random_state=42, max_iter=10000, C=0.8)

# 3. å»ºç«‹ Pipeline
pipeline = Pipeline([
    ('features', combined_features),
    ('classifier', model)
])

# æº–å‚™è³‡æ–™
X_input = df_train['clean_content'].astype(str)
y_input = df_train['sentiment']

# åˆ‡åˆ†é©—è­‰
X_tr, X_val, y_tr, y_val = train_test_split(X_input, y_input, test_size=0.1, random_state=42)

print("æ­£åœ¨è¨“ç·´ SVM (Feature Union ç‰ˆ)...")
pipeline.fit(X_tr, y_tr)

# é©—è­‰
val_preds = pipeline.predict(X_val)
val_f1 = f1_score(y_val, val_preds, average='macro')
print(f"Validation Macro F1 Score: {val_f1:.4f}")
print(classification_report(y_val, val_preds))

# --- å…¨è³‡æ–™è¨“ç·´èˆ‡æäº¤ ---
print("å…¨è³‡æ–™é‡æ–°è¨“ç·´ä¸­...")
pipeline.fit(X_input, y_input)
final_preds = pipeline.predict(df_test['clean_content'].astype(str))

submission = pd.DataFrame({
    'id': df_test['id'],
    'emotion': final_preds
})
submission.to_csv('submission_svm_union.csv', index=False)
print("submission_svm_union.csv å·²ç”Ÿæˆï¼é€™æ‡‰è©²æ˜¯æœ€ç©©å®šçš„ç‰ˆæœ¬ã€‚")
```
